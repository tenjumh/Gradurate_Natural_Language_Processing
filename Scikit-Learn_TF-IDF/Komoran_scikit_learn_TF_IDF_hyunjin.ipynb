```python
import json
from PyKomoran import *
from sklearn.feature_extraction.text import TfidfVectorizer

with open('C:/Users/fxk/PycharmProjects/tenjumh/Natural_Language_Processing/NLP_Gradurate/SpeechAct.json', 'r', encoding='utf-8') as json_file:
    corpus = json.load(json_file)

komoran = Komoran("EXP") 
POS_corpus = list()
```

```python
for i in corpus:
    if not corpus[i]:
        continue        
    for j in range(0, len(corpus[i])):
        POS_corpus.append(' '.join(komoran.get_morphes_by_tags(corpus[i][j][1], tag_list=['NNG', 'NNP', 'VV'])))
```

```python
POS_corpus
```

```python
tfidfvect = TfidfVectorizer(token_pattern='(?u)\\b\\w+\\b')
tfidfvect.fit_transform(POS_corpus)

with open('2019711974_김현진_TFIDF.txt', 'w') as f:
    for n in corpus:
        corpus_list = list()
        
        f.write('(대화{0})'.format(n))
        f.write('\n')
        
        if not corpus[n]:
            continue        
        for m in range(0, len(corpus[n])):
            corpus_list.append(' '.join(komoran.get_morphes_by_tags(corpus[n][m][1], tag_list=['NNG', 'NNP', 'VV'])))

        trans_list = tfidfvect.transform(corpus_list).toarray().tolist()

        for x in trans_list:
            f.write('\t'.join(str(tfv) for tfv in x))
            f.write('\n\n')
```
